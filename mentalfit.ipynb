```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in data.columns:
    if data[i].dtype=='object':   #transforming no-numerical values into numerical values
        data[i]=l.fit_transform(data[i])
    
```

```python
X= data.drop('mental_fitness',axis=1)
Y=data['mental_fitness']
```

```python
X
```

```python
#spiliting train & test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=.20,random_state=2)
```

```python
print("X_test :",X_test.shape)
print("X_train :",X_train.shape)

print("\n Y_test :",Y_test.shape)
print("Y_train :",Y_train.shape)
```

# Linear Regression algorithm

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lr=LinearRegression()

lr.fit(X_train,Y_train)


Y_train_predict =lr.predict(X_train)

mse=mean_squared_error(Y_train,Y_train_predict)

rmse=np.sqrt(mse)

r2=r2_score(Y_train,Y_train_predict)

print("Linear Regression performance on training dataset.......\n")

print(format(mse))
print(format(rmse))
print(format(r2))
```

```python
#training by using Random forest Algorithm
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_train,Y_train)

Y_train_predict =rf.predict(X_train)

mse=mean_squared_error(Y_train,Y_train_predict)

rmse=np.sqrt(mse)

r2=r2_score(Y_train,Y_train_predict)

print("Random Forest regressor model performance on training dataset.......\n")

print(format(mse))
print(format(rmse))
print(format(r2))
```

```python
Ytest_pred=lr.predict(X_test)
mse=mean_squared_error(Y_test,Ytest_pred)
rmse=np.sqrt(mse)

print("Linear Regression performance on test dataset.......\n")

print(format(mse))
print(format(rmse))
print(format(r2))

Ytest_pred
```

```python
Ytest_pred=rf.predict(X_test)
mse=mean_squared_error(Y_test,Ytest_pred)
rmse=np.sqrt(mse)

print("Random forest regressor performance on test dataset.......\n")

print(format(mse))
print(format(rmse))
print(format(r2))

Ytest_pred
```
